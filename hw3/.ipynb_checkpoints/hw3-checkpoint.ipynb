```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from polygon import RESTClient


api_key = "beBybSi8daPgsTp5yx5cHtHpYcrjp5Jq"
client = RESTClient(api_key=api_key)

# Define the currency pairs to use
ticker = "C:EURPLN"
date_from = "2023-04-12"
date_to = "2023-04-12"
fxdata = client.get_aggs(ticker=ticker, multiplier=1, timespan="hour", from_=date_from, to=date_to)
fxdata
```

```python
df = pd.DataFrame(fxdata)
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
```

```python
df['price_times_transactions'] = df['vwap'] * df['transactions']
```

```python
grouped_data = df.groupby(pd.Grouper(key='timestamp', freq='6H'))
```

```python

grouped_vwap = grouped_data['price_times_transactions'].sum() / grouped_data['transactions'].sum()
```

```python
grouped_liquidity = grouped_data['transactions'].sum() / 6
```

```python
# Combine the results into a new DataFrame
resampled_df = pd.DataFrame({
    'timestamp': grouped_vwap.index,
    'vwap': grouped_vwap.values,
    'liquidity': grouped_liquidity.values,
    'max': grouped_data['high'].max().values,
    'min': grouped_data['low'].min().values
})
```

```python
resampled_df
```

```python
# Calculate volatility and Keltner Channels
resampled_df['volatility'] = (resampled_df['max'] - resampled_df['min']) / resampled_df['vwap']
```

```python
# Calculate Keltner Channels
keltner_multiplier = 0.025 * np.arange(1, 101)
resampled_df['keltner_upper'] = list(np.outer(resampled_df['vwap'] + resampled_df['volatility'], keltner_multiplier))
resampled_df['keltner_lower'] = list(np.outer(resampled_df['vwap'] - resampled_df['volatility'], keltner_multiplier))
```

```python
resampled_df
```

```python
# Calculate FD for the first 6-hour interval
def count_crosses(prices, upper, lower):
    count = 0
    for i in range(1, len(prices)):
        if (prices[i - 1] < upper and prices[i] > upper) or (prices[i - 1] > lower and prices[i] < lower):
            count += 1
    return count
```

```python
first_interval_prices = df.loc[(df['timestamp'] >= resampled_df.iloc[0]['timestamp']) & (df['timestamp'] < resampled_df.iloc[1]['timestamp']), 'vwap']
```

```python
first_interval_prices
```

```python
first_interval_upper = resampled_df.iloc[0]['keltner_upper']
first_interval_lower = resampled_df.iloc[0]['keltner_lower']
first_interval_max = resampled_df.iloc[0]['max']
first_interval_min = resampled_df.iloc[0]['min']
```

```python
def count_crosses(prices, upper_list, lower_list):
    count = 0
    for upper, lower in zip(upper_list, lower_list):
        for i in range(1, len(prices)):
            if (prices[i - 1] < upper and prices[i] > upper) or (prices[i - 1] > lower and prices[i] < lower):
                count += 1
    return count

first_interval_upper = resampled_df.iloc[0]['keltner_upper']
first_interval_lower = resampled_df.iloc[0]['keltner_lower']
cross_count = count_crosses(first_interval_prices, first_interval_upper, first_interval_lower)
```

```python
resampled_df.loc[resampled_df.index[0], 'FD'] = cross_count / (first_interval_max - first_interval_min)
```
